```python
import os, subprocess

script = """
import os, sys
os.environ["CUDA_VISIBLE_DEVICES"] = ""
sys.path.insert(0, "/kaggle/working")

from pathlib import Path
import numpy as np
import pandas as pd
import json, hashlib, shutil
from datetime import datetime

ROOT = Path("/kaggle/working")
MAX_ROWS  = 200
MAX_PROFS = 5

print("=== Layer 1: Ingest ===")
df = pd.read_excel(ROOT / "edufeed_clean_data.xlsx", engine="openpyxl")
df = df.dropna(subset=["professor_name", "comments", "star_rating"])
df = df.drop_duplicates(subset=["professor_name", "comments"])
df = df.head(MAX_ROWS).reset_index(drop=True)
df.insert(0, "review_id", df.index.astype(str).str.zfill(6))
print("  " + str(len(df)) + " rows loaded")

print("=== Layer 2: Anonymise ===")
def sha256_token(v):
    return "PROF_" + hashlib.sha256(v.lower().strip().encode()).hexdigest()[:8].upper()
prof_map = {n: sha256_token(n) for n in df["professor_name"].dropna().unique()}
df["professor_token"] = df["professor_name"].map(prof_map)
df = df.drop(columns=["professor_name"])
df["comments"] = df["comments"].fillna("").astype(str)
print("  " + str(len(prof_map)) + " professors anonymised")

print("=== Layer 3: Embeddings (TF-IDF, no model download) ===")
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

texts = df["comments"].tolist()
print("  Fitting TF-IDF...")
tfidf = TfidfVectorizer(max_features=5000)
X = tfidf.fit_transform(texts)
print("  Reducing to 128 dims...")
svd = TruncatedSVD(n_components=128, random_state=42)
emb = svd.fit_transform(X).astype("float32")
# Normalize
norms = np.linalg.norm(emb, axis=1, keepdims=True)
emb = emb / np.where(norms == 0, 1, norms)
np.save(ROOT / "layer3_embeddings.npy", emb)
print("  embeddings shape: " + str(emb.shape))

def quick_sentiment(text):
    text = text.lower()
    pos = ["great","good","excellent","amazing","love","best","helpful","clear"]
    neg = ["bad","terrible","awful","boring","worst","confusing","hard","hate"]
    ps = sum(w in text for w in pos)
    ns = sum(w in text for w in neg)
    return "positive" if ps > ns else "negative" if ns > ps else "neutral"

df["sentiment"] = df["comments"].apply(quick_sentiment)
df["emotion"] = "neutral"
df.to_csv(ROOT / "layer3_enriched.csv", index=False)
print("  sentiment: " + str(df.sentiment.value_counts().to_dict()))

print("=== Layer 4: FAISS + Cards ===")
import faiss
dim = emb.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(emb)
faiss.write_index(index, str(ROOT / "layer4_faiss.index"))

prof_col = "professor_token"
professors = df[prof_col].unique()[:MAX_PROFS]
results = []
for prof in professors:
    prof_df = df[df[prof_col] == prof]
    reviews = prof_df["comments"].head(3).tolist()
    dept = prof_df["department_name"].mode().iloc[0] if "department_name" in prof_df.columns else "Unknown"
    avg_rating = round(prof_df["star_rating"].mean(), 2) if "star_rating" in prof_df.columns else "N/A"
    top_review = reviews[0][:100] if reviews else "N/A"
    results.append({
        "professor": prof, "department": dept, "avg_rating": avg_rating,
        "sentiments": prof_df["sentiment"].value_counts().to_dict(),
        "summary": "Based on " + str(len(prof_df)) + " reviews. Top: " + top_review,
        "strengths": ["See reviews"], "concerns": ["See reviews"],
    })
    print("  OK " + prof + " (" + str(len(prof_df)) + " reviews, " + str(avg_rating) + " stars)")

insights_path = str(ROOT / "insights.json")
with open(insights_path, "w") as f:
    json.dump(results, f, indent=2)
print("  " + str(len(results)) + " cards saved")

print("=== Layer 5: QA Check ===")
qa_status = "QA ready | professors: " + str(df[prof_col].nunique()) + " | cards: " + str(len(results))
rlhf = ROOT / "rlhf_feedback.jsonl"
if not rlhf.exists():
    rlhf.touch()
print("  " + qa_status)

print("=== Layer 6: Export ===")
outputs_dir = ROOT / "outputs"
outputs_dir.mkdir(exist_ok=True)
df.to_csv(outputs_dir / "predictions.csv", index=False)
shutil.copy2(insights_path, outputs_dir / "insights.json")
rating_avg = round(df["star_rating"].mean(), 3) if "star_rating" in df.columns else "N/A"
report = "# EduFeed Lite Report\\nGenerated: " + str(datetime.now()) + "\\nRows: " + str(len(df)) + "\\nAvg Rating: " + str(rating_avg) + "\\nStatus: COMPLETE\\n"
(outputs_dir / "report.md").write_text(report)
print("  predictions.csv, insights.json, report.md saved")
print("=== PIPELINE COMPLETE ===")
"""

with open("/kaggle/working/run_lite.py", "w") as f:
    f.write(script)

print("Running (no model download needed)...")
result = subprocess.run(
    ["python", "/kaggle/working/run_lite.py"],
    capture_output=False, text=True,
    cwd="/kaggle/working",
    env={**os.environ, "CUDA_VISIBLE_DEVICES": ""}
)
print("Return code:", result.returncode)
```

```python
import numpy as np
import pandas as pd
import faiss
import json
from pathlib import Path
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split

ROOT = Path("/kaggle/working")

df = pd.read_csv(ROOT / "outputs/predictions.csv")
emb = np.load(ROOT / "layer3_embeddings.npy")

# ── 1. Sentiment Accuracy ─────────────────────────────────────────
print("=" * 55)
print("1. SENTIMENT ACCURACY")
print("=" * 55)

# Use original sentiment_label as ground truth (from raw data)
if "sentiment_label" in df.columns:
    # Map ground truth to match our labels
    label_map = {"positive": "positive", "negative": "negative", "neutral": "neutral",
                 "Positive": "positive", "Negative": "negative", "Neutral": "neutral",
                 1: "positive", 0: "neutral", -1: "negative"}
    df["gt_sentiment"] = df["sentiment_label"].map(label_map).fillna("neutral")
    y_true = df["gt_sentiment"]
    y_pred = df["sentiment"]

    acc = accuracy_score(y_true, y_pred)
    print(f"Accuracy : {acc:.3f} ({acc*100:.1f}%)")
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred, labels=["positive", "neutral", "negative"]))
else:
    print("No ground truth 'sentiment_label' column found.")
    print("Sentiment distribution in predictions:")
    print(df["sentiment"].value_counts())

# ── 2. Embedding Quality (FAISS retrieval) ────────────────────────
print("\n" + "=" * 55)
print("2. EMBEDDING / RETRIEVAL QUALITY")
print("=" * 55)

index = faiss.read_index(str(ROOT / "layer4_faiss.index"))

# Test: for each review, find top-5 neighbors
# Check if same professor appears in top-5 (proxy for retrieval quality)
prof_col = "professor_token"
hit_at_1 = 0
hit_at_5 = 0
sample_size = min(100, len(df))
sample_idx = np.random.choice(len(df), sample_size, replace=False)

for i in sample_idx:
    q_vec = emb[i].reshape(1, -1).astype("float32")
    D, I = index.search(q_vec, k=6)  # k=6 because idx 0 is itself
    neighbors = [j for j in I[0] if j != i][:5]
    true_prof = df.iloc[i][prof_col]
    neighbor_profs = df.iloc[neighbors][prof_col].tolist()
    if neighbor_profs[0] == true_prof:
        hit_at_1 += 1
    if true_prof in neighbor_profs:
        hit_at_5 += 1

print(f"Same-professor retrieval Hit@1 : {hit_at_1/sample_size:.3f} ({hit_at_1/sample_size*100:.1f}%)")
print(f"Same-professor retrieval Hit@5 : {hit_at_5/sample_size:.3f} ({hit_at_5/sample_size*100:.1f}%)")
print(f"(How often top-K neighbors belong to the same professor)")

# ── 3. Rating Distribution ────────────────────────────────────────
print("\n" + "=" * 55)
print("3. RATING DISTRIBUTION")
print("=" * 55)
print(df["star_rating"].describe().round(3))
bins = [1, 2, 3, 4, 5]
labels = ["1★", "2★", "3★", "4★", "5★"]
df["rating_bin"] = pd.cut(df["star_rating"], bins=5, labels=labels)
print("\nRating bins:")
print(df["rating_bin"].value_counts().sort_index())

# ── 4. Professor Card Summary ─────────────────────────────────────
print("\n" + "=" * 55)
print("4. PROFESSOR INSIGHT CARDS")
print("=" * 55)
with open(ROOT / "outputs/insights.json") as f:
    cards = json.load(f)

for card in cards:
    print(f"\n{card['professor']} | {card['department']} | {card['avg_rating']}★")
    print(f"  Sentiments : {card['sentiments']}")
    print(f"  Summary    : {card['summary'][:120]}")

# ── 5. Overall Pipeline Score ─────────────────────────────────────
print("\n" + "=" * 55)
print("5. OVERALL PIPELINE SUMMARY")
print("=" * 55)
print(f"Total reviews processed : {len(df)}")
print(f"Professors anonymised   : {df[prof_col].nunique()}")
print(f"Insight cards generated : {len(cards)}")
print(f"Embedding dimensions    : {emb.shape[1]}")
print(f"FAISS index size        : {index.ntotal}")
print(f"Avg star rating         : {df['star_rating'].mean():.3f}")
print(f"Sentiment breakdown     : {df['sentiment'].value_counts().to_dict()}")
if "sentiment_label" in df.columns:
    print(f"Sentiment accuracy      : {acc*100:.1f}%")
print("=" * 55)
```

```python
import os, subprocess

script = """
import os, sys
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
sys.path.insert(0, "/kaggle/working")

from pathlib import Path
import numpy as np
import pandas as pd
import json, hashlib, shutil
from datetime import datetime
import torch

ROOT = Path("/kaggle/working")
MAX_PROFS = 20

print("GPU available:", torch.cuda.is_available())
print("GPU name:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")

print("=== Layer 1: Ingest ===")
df = pd.read_excel(ROOT / "edufeed_clean_data.xlsx", engine="openpyxl")
df = df.dropna(subset=["professor_name", "comments", "star_rating"])
df = df.drop_duplicates(subset=["professor_name", "comments"])
df = df.reset_index(drop=True)
df.insert(0, "review_id", df.index.astype(str).str.zfill(6))
print("  " + str(len(df)) + " rows loaded")

print("=== Layer 2: Anonymise ===")
def sha256_token(v):
    return "PROF_" + hashlib.sha256(v.lower().strip().encode()).hexdigest()[:8].upper()
prof_map = {n: sha256_token(n) for n in df["professor_name"].dropna().unique()}
df["professor_token"] = df["professor_name"].map(prof_map)
df = df.drop(columns=["professor_name"])
df["comments"] = df["comments"].fillna("").astype(str)
print("  " + str(len(prof_map)) + " professors anonymised")

print("=== Layer 3: SBERT Embeddings (GPU) ===")
from sentence_transformers import SentenceTransformer
device = "cuda" if torch.cuda.is_available() else "cpu"
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2", device=device)
texts = df["comments"].tolist()
emb = model.encode(texts, batch_size=256, show_progress_bar=True, normalize_embeddings=True)
np.save(ROOT / "layer3_embeddings.npy", emb)
print("  embeddings shape: " + str(emb.shape))

print("=== Layer 3b: DeBERTa Zero-shot Sentiment (GPU) ===")
from transformers import pipeline as hf_pipeline
clf = hf_pipeline(
    "zero-shot-classification",
    model="cross-encoder/nli-deberta-v3-base",
    device=0,
    batch_size=64
)
SENTIMENT_LABELS = ["positive", "negative", "neutral"]
sentiments = []
for i in range(0, len(texts), 64):
    batch = texts[i:i+64]
    results = clf(batch, SENTIMENT_LABELS)
    sentiments.extend([r["labels"][0] for r in results])
    if i % 500 == 0:
        print("  Progress: " + str(i) + "/" + str(len(texts)))

df["sentiment"] = sentiments
df["emotion"] = "neutral"
df.to_csv(ROOT / "layer3_enriched.csv", index=False)
print("  sentiment: " + str(df.sentiment.value_counts().to_dict()))

print("=== Layer 4: FAISS + Cards ===")
import faiss
dim = emb.shape[1]
index = faiss.IndexHNSWFlat(dim, 32)
index.hnsw.efConstruction = 200
index.add(emb.astype("float32"))
faiss.write_index(index, str(ROOT / "layer4_faiss.index"))

prof_col = "professor_token"
professors = df[prof_col].unique()[:MAX_PROFS]
results = []
for prof in professors:
    prof_df = df[df[prof_col] == prof]
    reviews = prof_df["comments"].head(5).tolist()
    dept = prof_df["department_name"].mode().iloc[0] if "department_name" in prof_df.columns else "Unknown"
    avg_rating = round(prof_df["star_rating"].mean(), 2) if "star_rating" in prof_df.columns else "N/A"
    top_review = reviews[0][:150] if reviews else "N/A"
    results.append({
        "professor": prof, "department": dept, "avg_rating": avg_rating,
        "sentiments": prof_df["sentiment"].value_counts().to_dict(),
        "summary": "Based on " + str(len(prof_df)) + " reviews. Top: " + top_review,
        "strengths": ["See reviews"], "concerns": ["See reviews"],
    })
    print("  OK " + prof + " (" + str(len(prof_df)) + " reviews, " + str(avg_rating) + " stars)")

insights_path = str(ROOT / "insights.json")
with open(insights_path, "w") as f:
    json.dump(results, f, indent=2)
print("  " + str(len(results)) + " cards saved")

print("=== Layer 5: QA Check ===")
qa_status = "QA ready | professors: " + str(df[prof_col].nunique()) + " | cards: " + str(len(results))
(ROOT / "rlhf_feedback.jsonl").touch()
print("  " + qa_status)

print("=== Layer 6: Export ===")
outputs_dir = ROOT / "outputs"
outputs_dir.mkdir(exist_ok=True)
df.to_csv(outputs_dir / "predictions.csv", index=False)
shutil.copy2(insights_path, outputs_dir / "insights.json")
rating_avg = round(df["star_rating"].mean(), 3) if "star_rating" in df.columns else "N/A"
report = "# EduFeed Full Report\\nGenerated: " + str(datetime.now()) + "\\nRows: " + str(len(df)) + "\\nAvg Rating: " + str(rating_avg) + "\\nStatus: COMPLETE\\n"
(outputs_dir / "report.md").write_text(report)
print("  predictions.csv, insights.json, report.md saved")
print("=== PIPELINE COMPLETE ===")
"""

with open("/kaggle/working/run_full.py", "w") as f:
    f.write(script)

print("Running full pipeline on T4 GPU...")
result = subprocess.run(
    ["python", "/kaggle/working/run_full.py"],
    capture_output=False, text=True,
    cwd="/kaggle/working",
    env={**os.environ, "CUDA_VISIBLE_DEVICES": "0"}
)
print("Return code:", result.returncode)
```

```python
import numpy as np
import pandas as pd
import faiss
import json
from pathlib import Path
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

ROOT = Path("/kaggle/working")

df  = pd.read_csv(ROOT / "outputs/predictions.csv")
emb = np.load(ROOT / "layer3_embeddings.npy")
index = faiss.read_index(str(ROOT / "layer4_faiss.index"))

with open(ROOT / "outputs/insights.json") as f:
    cards = json.load(f)

# ── 1. Sentiment Accuracy ─────────────────────────────────────────
print("=" * 55)
print("1. SENTIMENT ACCURACY")
print("=" * 55)

if "sentiment_label" in df.columns:
    label_map = {
        "positive": "positive", "negative": "negative", "neutral": "neutral",
        "Positive": "positive", "Negative": "negative", "Neutral": "neutral",
        1: "positive", 0: "neutral", -1: "negative"
    }
    df["gt"] = df["sentiment_label"].map(label_map).fillna("neutral")
    y_true = df["gt"]
    y_pred = df["sentiment"]
    acc = accuracy_score(y_true, y_pred)
    print("Accuracy : " + str(round(acc*100, 2)) + "%")
    print(classification_report(y_true, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred, labels=["positive", "neutral", "negative"]))
else:
    print("Sentiment distribution (no ground truth column found):")
    print(df["sentiment"].value_counts())
    print("\nTip: if your Excel has a sentiment column with different name, tell me!")

# ── 2. Retrieval Quality ──────────────────────────────────────────
print("\n" + "=" * 55)
print("2. RETRIEVAL QUALITY (FAISS)")
print("=" * 55)

prof_col = "professor_token"
hit1 = hit5 = 0
sample = min(200, len(df))
idx_sample = np.random.choice(len(df), sample, replace=False)

for i in idx_sample:
    q = emb[i].reshape(1,-1).astype("float32")
    _, I = index.search(q, k=6)
    neighbors = [j for j in I[0] if j != i][:5]
    true_prof = df.iloc[i][prof_col]
    neighbor_profs = df.iloc[neighbors][prof_col].tolist()
    if neighbor_profs[0] == true_prof: hit1 += 1
    if true_prof in neighbor_profs:    hit5 += 1

print("Hit@1 : " + str(round(hit1/sample*100, 1)) + "%")
print("Hit@5 : " + str(round(hit5/sample*100, 1)) + "%")

# ── 3. Overall Summary ────────────────────────────────────────────
print("\n" + "=" * 55)
print("3. OVERALL SUMMARY")
print("=" * 55)
print("Total reviews     : " + str(len(df)))
print("Professors        : " + str(df[prof_col].nunique()))
print("Insight cards     : " + str(len(cards)))
print("Embedding dims    : " + str(emb.shape[1]))
print("Avg star rating   : " + str(round(df["star_rating"].mean(), 3)))
print("Sentiment dist    : " + str(df["sentiment"].value_counts().to_dict()))
print("=" * 55)
```
